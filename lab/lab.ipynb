```python
import os
import time

import gymnasium
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.vec_env import DummyVecEnv
```

```python
environment_name = "CartPole-v1"
```

```python
env = gymnasium.make(environment_name)
```

```python
env.reset()
```

```python
env.action_space
```

```python
env.observation_space
```

```python
episodes = 10
for episode in range(1, episodes):
    env.reset()
    terminated = False
    truncated = False
    score = 0

    while not terminated and not truncated:
        action = env.action_space.sample()
        n_state, reward, terminated, truncated, info = env.step(action)
        score += reward
    print(f"Episode: {episode}  Score {score}")
env.close()
time.sleep(10)
```

```python
log_path = os.path.join("training", "logs")
model_path = os.path.join("training", "models")
```

```python
log_path
```

```python
env = gymnasium.make(environment_name, render_mode="human")
env = DummyVecEnv([lambda: env])
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=20000)
```

```python
PPO_Path = os.path.join(model_path,"PPO_Model_CartPole")
```

```python
model.save(PPO_Path)
```

```python
del model
```

```python
model = PPO.load(PPO_Path, env)
```

```python
evaluate_policy(model, env, 1, render=True)
```

```python
episodes = 5
for episode in range(1, episodes):
    obs = env.reset()
    done = False
    score = 0

    while not done:
        action, _ = model.predict(obs)
        obs, reward, done, info = env.step(action)
        score += reward
    print(f"Episode: {episode}  Score {score}")
env.close()
```

```python
from stable_baselines3.common.callbacks import (
    EvalCallback,
    StopTrainingOnRewardThreshold,
)
```

```python
env = gymnasium.make(environment_name)
```

```python
save_path = os.path.join("Training", "Models")
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200, verbose=1)
eval_callback = EvalCallback(
    eval_env=env,
    callback_on_new_best=stop_callback,
    eval_freq=10000,
    best_model_save_path=save_path,
    verbose=1
)
```

```python
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```

```python
net_arch = dict(pi=[128,128,128,128], vf=[128,128,128,128])
```

```python
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log=log_path,policy_kwargs={"net_arch":net_arch})
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```
